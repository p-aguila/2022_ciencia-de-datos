## Matriz de covarianza

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("seaborn-dark")
from sklearn.preprocessing import StandardScaler # z-score
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.pairplot(iris, hue='species')
plt.show()
```

```python
sns.pairplot(iris, height=5, aspect=0.8, kind='reg')
plt.show()
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(iris[['sepal_length', 
                                    'sepal_width', 
                                    'petal_length', 
                                    'petal_width']])
```

```python
scaled.shape
```

```python
scaled.T.shape # transponer matrix
```

```python
covariance_matrix = np.cov(scaled.T)
covariance_matrix # esta matriz se relaciona con el grafico de arriba
```

```python
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
heat_map = sns.heatmap(covariance_matrix,
                      cbar=True,
                      annot=True,
                      square=True,
                      annot_kws={'size':12},
                      yticklabels=['sepal_length',
                                   'sepal_width', 
                                   'petal_length', 
                                   'petal_width'],
                      xticklabels=['sepal_length',
                                   'sepal_width', 
                                   'petal_length', 
                                   'petal_width'])
```

```python
sns.heatmap(iris.corr(), annot=True)
plt.show()
```
