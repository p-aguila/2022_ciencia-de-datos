```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits
```

```python
# esto no es un dataframe de pandas, contiene la data, los pixeles de las imagenes, entre otros
digitos = load_digits()
# array con la data
datos_digitos = digitos.data
print(datos_digitos.shape)
```

```python
np.random.seed(10)
k = 10 # numero de clusters
kmeans = KMeans(n_clusters=k, init='random')
```

```python
kmeans.fit(datos_digitos)
```

```python
kmeans_pred = kmeans.predict(datos_digitos)
```

```python
kmeans_pred.shape
```

```python
kmeans_pred
```

```python
for i in range(0, k):
    fila = np.where(kmeans_pred == i)[0] # esto es un array dentro de un array
    num = fila.shape[0]
    r = np.floor(num/10.)
    
    print("Cluster " + str(i))
    print("Numero de elementos " + str(num))
    
    plt.figure(figsize=(10,10))
    for j in range(1, num):
        plt.subplot(r + 1, 10, j)
        imagen = datos_digitos[fila[j],]
        imagen = imagen.reshape(8,8)
        plt.imshow(imagen, cmap=plt.cm.gray)
        plt.axis('off')
    plt.show()
```
