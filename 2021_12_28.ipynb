## GIT

Comandos GIT
1. `cd`: arrastrar la carpeta en la que quiero trabajar
2. `pwd`: ruta en la que estoy
3. `ls`: ver archivos dentro de la carpeta
4. `git init`: inicializar repositorio
5. `git status`: archivos listos para entrar a produccion (archivos en rojo)
6. `git add .`: para subir los archivos
7. `git status`: comprobar si se subieron (archivos en verde) 
8. `git commit -m "Repositorio curso Ciencia de Datos"`: describir los cambios que estoy haciendo con el commit. 
9. `git branch -M main` 
10. `git remote add origin https://github.com/p-aguila/2022_ciencia-de-datos.git`
11. `git push -u origin main`

## Tarea:
- elegir un modelo de auto
- construir un df
- hacer un analisis con un plot, relacion precio/año de produccion
- conclusion/observaciones

```python
# elegir un modelo de auto
# elijo "Mercedes-Benz"
df.groupby('manufacturer_name')['manufacturer_name'].count()
```

```python
# construir un df del manufacturer_name y model_name
mercedes_df = df[(df['manufacturer_name'] == 'Mercedes-Benz') & 
                 (df['model_name'] == 'Sprinter')]
mercedes_df
```

```python
# hacer un analisis con un plot, relacion precio/año de produccion
sns.histplot(mercedes_df, x = 'price_usd', hue = 'year_produced')
plt.show()
```

```python
# Buscar datos atípicos
mercedes_df[(mercedes_df['price_usd'] >= 40000)]
```

```python
# Buscar datos atípicos
mercedes_df[(mercedes_df['price_usd'] <= 2500)]
```

**Conclusiones**:

1. A diferencia de los autos Audi Q7, se observa que los vehículos Mercedes-Benz Sprinter tienen un mayor rango de años en su fabricación (desde 1986 a 2017).
2. Hay una gran concentración de vehículos cuyos valores rondan entre los 5 mil y 10 mil dólares.
3. Se observa un posible dato atípico a la derecha (auto con precio mayor a 40.000 USD).
