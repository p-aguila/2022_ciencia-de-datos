BATCH GRADIENT DESCENT

```python
from sklearn.datasets import make_regression
```

```python
n_muestra = 1000
n_variables = 2
```

```python
X, y, coeficientes_objetivos = make_regression(n_samples=n_muestra, 
                                               n_features=n_variables,
                                               coef=True)
```

```python
print(X.shape, y.shape, coeficientes_objetivos.shape)
```

```python
coeficientes_objetivos
```

```python
def prediccio_batch(coeficientes, X):
    return coeficientes @ X.T
```

```python
y_predicciones = prediccio_batch(coeficientes_objetivos, X)
y_predicciones[:5]
```

```python
# comprobando que la prediccion es correcta (ojo, no es el error), 
# solo se comprueba que la funcion es correcta
y[:5] - y_predicciones[:5]
```

```python
# claculamos el error cuadratico medio
def error_batch(y_prediccion, y_real):
    n = y_prediccion.shape[0]
    return (np.sum(y_prediccion - y_real)**2/2*n)
```

```python
def derivada_error_batch(y_prediccion, y_real, x):
    n = y_prediccion.shape[0]
    return np.sum((y_prediccion - y_real)*x/n)
```

```python
coeficientes = np.random.random((X.shape[1], ))
coeficientes
```

```python
def descenso_gradiente_batch(coef, X, y):
    y_predic = prediccio_batch(coef, X)
    for i in range(coef.shape[0]):
        coef[i] = coef[i] - alfa * derivada_error_batch(y_predic, y, X[:,i])
    error = error_batch(y_predic, y)
    return coef, error
```

```python
coef_iteraciones = []
error_iteraciones = []

NUM_ITERACIONES = 300
coeficientes = np.random.random((X.shape[1],))
error = error_batch(coeficientes,X)
for i in range(NUM_ITERACIONES):
    coef_iteraciones.append(coeficientes.copy())
    error_iteraciones.append(error)
    coeficientes, error = descenso_gradiente_batch(coeficientes, X, y)
coef_iteraciones = np.array(coef_iteraciones)
```
