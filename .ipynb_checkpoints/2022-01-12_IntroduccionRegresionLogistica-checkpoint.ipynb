```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use("seaborn-dark")
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

- la regresion logistica se basa en predecir valores binarios
- el problema con este dataset es que tenemos tres especies
- para este ejercicio vamos a quitar una de las especies

```python
new_iris = iris[iris['species']!='setosa']
```

```python
sns.pairplot(new_iris, hue='species')
plt.show()
```

```python
X = new_iris.drop('species', axis=1)
y = new_iris['species'] # lo que vamos a predecir
```

```python
from sklearn.model_selection import train_test_split
```

```python
# ?train_test_split
# sirve para ver la documentacion oficial 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42) 
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reglog_model = LogisticRegression()
```

```python
# entrenar el modelo
reglog_model.fit(X_train, y_train)
```

```python
# predicciones hechas en base a X_test
predicciones = reglog_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, predicciones)
```
