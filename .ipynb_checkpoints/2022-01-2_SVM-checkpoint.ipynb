```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
df = load_iris()
```

```python
print(df.DESCR)
```

```python
from IPython.display import Image
Image('iris.png')
```

```python
iris = pd.DataFrame(df.data, columns=df.feature_names)
iris['objetivo'] = df.target
iris.head()
```

```python
var_indep = iris.drop('objetivo', axis=1).columns
var_indep
```

```python
iris_X = iris[var_indep]
iris_y = iris['objetivo']
```

```python
from sklearn.model_selection import train_test_split
iris_X_train, iris_X_test, iris_y_train, iris_y_test = train_test_split(iris_X, 
                                                                        iris_y, 
                                                                        test_size=0.2)
```

```python
from sklearn.svm import SVC
estimador_svm = SVC()
```

```python
estimador_svm.fit(iris_X_train, iris_y_train)
```

```python
estimador_svm.predict(iris_X_test)
```

```python
estimador_svm.support_vectors_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
estimador_svm_lineal  = SVC(kernel='linear')
```

```python
estimador_svm_lineal.fit(iris_X_train, iris_y_train)
```

```python
Image('kernel_lineal.png')
```

```python
Xp = df.data[:,:2]
yp = df.target
estimador_svm_lineal.fit(Xp, yp)
```

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-dark")
```

```python
plot_decision_regions(Xp, yp, clf=estimador_svm_lineal)
```

```python
estimador_svm_polinomial = SVC(kernel='poly', degree=6)
estimador_svm_polinomial.fit(Xp, yp)
plot_decision_regions(Xp, yp, clf=estimador_svm_polinomial)
```
