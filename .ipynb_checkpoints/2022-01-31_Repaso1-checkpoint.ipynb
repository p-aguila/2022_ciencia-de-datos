```python
import pandas as pd
```

```python
filename = '../Ejercicios_Propios/datasets/pima-indians-diabetes.csv'
nombres = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
df = pd.read_csv(filename, names=nombres)
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("seaborn-dark")
fig = plt.figure(figsize=(10, 10))
ax = fig.gca()
df.hist(ax=ax)
plt.show()
```

```python
# graficos boxplot
fig = plt.figure(figsize=(16, 16))
ax = fig.gca()
df.plot(ax=ax, kind='box', subplots=True, layout=(3, 3))
plt.show()
```

```python
# matriz de correlacion
correlaciones = df.corr()
correlaciones
```

```python
import numpy as np
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlaciones, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0, 9, 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(nombres)
ax.set_yticklabels(nombres)
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 10))
ax = sns.heatmap(correlaciones, vmax=1, square=True, annot=True, cmap='viridis')
```

```python
from pandas.plotting import scatter_matrix
plt.rcParams['figure.figsize'] = [15,15]
scatter_matrix(df)
plt.show()
```

```python
sns.pairplot(df, hue='class', diag_kind='hist')
plt.show()
```
