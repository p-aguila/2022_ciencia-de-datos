## Ver correlaci√≥n entre variables

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("seaborn-dark")

iris = sns.load_dataset('iris')
iris.head()
```

```python
# scatterplot
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
plt.grid()
plt.show()
```

```python
# joint plot
sns.jointplot(data=iris, x='sepal_length', y='petal_length', hue='species')
plt.show()
```

```python
# box plot
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.show()
```

```python
# bar plot
sns.barplot(data=iris, x='species', y='sepal_length')
plt.show()
```

```python
# esto no se hace, el hue solo pinta, no filtra:
# joint plot
sns.jointplot(data=iris, 
              x='petal_length', 
              y='sepal_length', 
              hue='species' == 'setosa')
plt.show()
```

```python
# generando el df de setosa
iris_setosa = iris[iris['species'] == 'setosa']

iris_setosa
```

```python
# joint plot
sns.jointplot(data=iris_setosa, x='petal_length', y='sepal_length')
plt.show()
```
