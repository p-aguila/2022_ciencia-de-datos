```python
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
datos_cancer = datasets.load_breast_cancer()
```

```python
df = pd.DataFrame(datos_cancer['data'], columns=datos_cancer.feature_names)
df.head()
```

```python
df['objetivo'] = datos_cancer.target
df
```

```python
df.shape
```

```python
df.objetivo.value_counts(True)
```

```python
df.objetivo.value_counts(False)
```

```python
# sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
train_df, test_df = train_test_split(df, test_size=0.3)
variables_independientes = datos_cancer.feature_names
variable_objetivo = 'objetivo'
```

```python
train_df.shape
```

```python
test_df.shape
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("seaborn-dark")
```

```python
# aplicar reg lineal a una rel logistica, solo para ver como queda
columna_independiente = 'worst area'

plt.plot(train_df[columna_independiente], train_df.objetivo, '.b')
plt.xlabel('Peor area encontrada')
plt.ylabel('Objetivo (Maligno/Benigno)')
plt.show()
```

```python
modelo_ols = LinearRegression()
modelo_ols.fit(train_df[[columna_independiente]], train_df[variable_objetivo])
predicciones = modelo_ols.predict(test_df[[columna_independiente]])
```

```python
predicciones[:20]
```

```python
plt.plot(test_df[columna_independiente], test_df.objetivo, '.b')
plt.plot(test_df[columna_independiente], predicciones, '.r')
plt.xlabel('Peor area encontrada')
plt.ylabel('Objetivo (Maligno/Benigno)')
plt.show()
```

```python
# reg log
from sklearn.linear_model import LogisticRegression 
```

```python
X = df[variables_independientes]
y = df[variable_objetivo]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
reg_log = LogisticRegression()
reg_log.fit(X_train, y_train)
predicciones = reg_log.predict(X_test)

predicciones[:10]
```

```python
predicciones_probabilidades = reg_log.predict_proba(X_test)
predicciones_probabilidades[:10]
```

```python
plt.hist(predicciones_probabilidades)
```

```python
probs_df = pd.DataFrame(predicciones_probabilidades)
probs_df[:10]
```

```python
X = X_test.reset_index().copy() # resetear indices
X['objetivo'] = y_test.to_list() # to_list: se ordenan los indices para el matching
X['prediccion'] = predicciones
X = pd.concat([X, probs_df], axis=1)
X[['objetivo', 'prediccion', 0, 1]].head(30)
```

```python
probabilidades_ordenadas = pd.Series(X[1]).sort_values(ascending=True)
probabilidades_ordenadas = probabilidades_ordenadas.reset_index().copy()
```

```python
plt.plot(probabilidades_ordenadas[1], '.r')
plt.xlabel('Casos')
plt.ylabel('Probabilidad de pertenecer a categoria 1')
plt.show()
```
